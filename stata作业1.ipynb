```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 50)
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
df=pd.read_stata('cgss样本.dta',convert_categoricals=True)
df.head()
```

```python
def fre(series):
    tabulate=pd.concat([series.value_counts(),
           series.value_counts()/len(series)],axis=1)
    tabulate.columns=['freq.','per.']
    tabulate.loc['total']=tabulate.apply(lambda x:x.sum())
    return tabulate
```

```python
#性别 a2 男/女
fre(df['a2'])
```

```python
#民族 a4 汉族/少数民族
def nation_trans(i):
    if i == '汉':
        return '汉族'
    else:
        return '少数民族'
fre(df['a4'].apply(nation_trans))
```

```python
#户籍身份 a18 城市/农村
d={'农业户口':'农村','非农业户口':'城市','居民户口（以前是非农业户口）':'城市','居民户口（以前是农业户口）':'城市','军籍':'其他'}
fre(df['a18'].replace(d))
```
