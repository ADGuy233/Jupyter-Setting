```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 100)
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
df=pd.read_stata('cgss样本.dta',convert_categoricals=True)
df.head()
```

```python
def fre(series):
    tabulate=pd.concat([series.value_counts(),
           series.value_counts()/len(series)],axis=1)
    tabulate.columns=['freq.','per.']
    tabulate.loc['total']=tabulate.apply(lambda x:x.sum())
    return tabulate
```

```python
#性别 a2 男/女
fre(df['a2'])
```

```python
#民族 a4 汉族/少数民族
def nation_trans(i):
    if i == '汉':
        return '汉族'
    else:
        return '少数民族'
fre(df['a4'].apply(nation_trans))
```

```python
#户籍身份 a18 城市/农村
d_register={'农业户口':'农村','非农业户口':'城市','居民户口（以前是非农业户口）':'城市','居民户口（以前是农业户口）':'城市','军籍':'其他'}
fre(df['a18'].replace(d_register))
```

```python
#受教育程度 a7a 需剔除仍在就读的个案
fre(df.loc[df['a7b']!='正在读','a7a'])
```

```python
#收入水平 a8a 自己拟定标准进行分组
df['income'] = pd.to_numeric(df['a8a'],errors='coerce')
print(fre(df['a8a']))
income_df=pd.cut(df['income'],bins=[0,5000,30000,80000,1000000],
       labels=['贫穷','温饱','小康','富裕'])
fre(income_df)
```

```python
#父亲政治面貌 a89c 是否为党员
def party_trans(i):
    if i == '共产党员':
        return '党员'
    else:
        return '非党员'
fre(df['a89c'].apply(party_trans))
```

```python
#父亲的职业类型 a89g 自行处理
d_a89g={'无单位/自雇（包括个体户）':'企业',
        '事业单位':'体制','党政机关':'体制','军队':'体制','社会团体、居/村委会':'体制',
        '不知道':np.nan,'拒绝回答':np.nan
       }
fre(df['a89g'].replace(d_a89g))
```

```python
#父亲的受教育程度 a89b 自行处理
d_a89b={'小学':'小学及以下','私塾、扫盲班':'小学及以下','没有受过任何教育':'小学及以下','中专':'初中',
      '技校':'高中','研究生及以上':'大学及以上','不知道':np.nan,'拒绝回答':np.nan}
def edu_trans(i):
    if '高中' in i :
        return '高中'
    elif '大学' in i :
        return '大学及以上'
    else:
        return i
fre(df['a89b'].apply(edu_trans).replace(d_a89b))
```

```python
#兄弟姐妹数量 d21需自行计算
fre(df.loc[:,'d21_1':'d21_4'].sum(axis=1))
```

```python
df['class'] = pd.to_numeric(df['a43b'],errors='coerce')
fre(pd.cut(df['class'],bins=[0,3,6,11],labels=['1-3','4-6','6-10']))
```

```python
fre(df['a51'])
```

```python
#2.对样本的受教育年限（根据教育程度自行转化）分性别、分城乡进行描述统计。（1）基本情况的测量，如均值、标准差等
d_eduage={'小学':6,'私塾、扫盲班':6,'没有受过任何教育':0,'中专':9,'初中':9,
      '技校':12,'研究生及以上':18,'不知道':np.nan,'拒绝回答':np.nan,'其他':np.nan}
def edu_agetrans(i):
    if '高中' in i :
        return 12
    elif '大学' in i :
        return 16
    else:
        return i
df['eduage']=df['a7a'].apply(edu_agetrans).replace(d_eduage)
print(df[['eduage','a2']].groupby('a2').describe().T)
df[['eduage','a18']].groupby('a18').describe().T
```

```python
pd.concat([fre(df['a2']),fre(df['a4'].apply(nation_trans)),
           fre(df['a18'].replace(d_register)),
           fre(df.loc[df['a7b']!='正在读','a7a']),
           fre(income_df),fre(df['a89c'].apply(party_trans)),
           fre(df['a89g'].replace(d_a89g)),
           fre(df['a89b'].apply(edu_trans).replace(d_a89b)),
           fre(df.loc[:,'d21_1':'d21_4'].sum(axis=1)),
          fre(pd.cut(df['class'],bins=[0,3,6,11],labels=['1-3','4-6','6-10'])),
          fre(df['a51'])]).to_excel('stata作业.xlsx')
```
